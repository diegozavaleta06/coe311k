```python
import numpy
def factorial (x):
    upval = x
    for z in range(1,x):
        upval = upval * (x-z)
    print(upval)
    return(upval)





def sin_number(terms):
    listx = []
    for i in range(terms):
        flipper = 2
        expy = 3
        firstexcp = 0
        for k in numpy.arange(0,10,0.1):
            if firstexcp ==0:
                sum = k
            term = (k**expy) / factorial(expy)
            if flipper%2 == 0:
                 sum = sum -  term
            if flipper%2 == 1:
                sum = sum + term   
            expy += 2
            flipper += 1
            real_sin = numpy.sin(k)
            abs_error = real_sin - term
            listx.append(abs_error)
            firstexcp += 1
    return(listx)
   
```
