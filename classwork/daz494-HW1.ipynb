```python
import math
import random
#This ^ imports the functions we need for the function to work.

def pi_counter (num_samples):
    counter = 0
    #this counter is the variable to see how many hits in the area land
    for s in range(0,num_samples):
        value = random.uniform(-1,1)
        value2 = random.uniform(-1,1)
        distance = value*value + value2*value2
        #creates two random numbers and add the sum of squares
        if distance <= 1:
            counter += 1
            #adds to the counter if it meets the condition
    return(4 * (counter/num_samples))  
```

```python
import matplotlib.pyplot as plt #imports what we need to start plotting
x = range(10,1000,20)
#creates ^ the range for what we need in increments of 20
absvec = []
relyvec = []
# The two vectors we need to store the data in list structure ^
for i in x:
    dubby = pi_counter(i) #called the pi_counter function with the size that we change
    abs_error = abs(math.pi - dubby)
    rely_error = abs((math.pi - dubby)/math.pi)
    absvec.append(abs_error)
    relyvec.append(rely_error)
    #This ^ calculates the two different error values and then adds them to their respective lists
plt.plot(x, absvec, "r--", label="Absolute Error")
plt.plot(x, relyvec, "b", label="Relative Error")
plt.legend()
plt.show()
#This plots each different list and has them be two different colors, adds a legend and then shows it.
```
