```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 50)

plt.subplot(4, 1, 1) 
plt.plot(x, 0.5*x + 1, 'r')

plt.subplot(4,1,2)
plt.scatter(x,np.sin(x),10)

plt.subplot(4,1,3)
y = np.random.randn(1000)
plt.hist(y)

plt.subplot(4,1,4)
l = np.random.rand(50)
m = np.random.rand(50)
size = np.random.rand(50) * 50
color = np.random.rand(50)
plt.scatter(l, m, size, color)
plt.colorbar()


plt.show()
```
