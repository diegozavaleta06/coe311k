```python
import numpy
negative =3
sum = 1-(1/3)+(1/5)

for i in range(3,5000):
    fraction = 1/(2*i + 1)
    if negative%2 == 0:
        sum = sum + fraction
    if negative%2 == 1:
        sum = sum - fraction
    testpi = 4*sum
    if i%100 == 0:
        print(f"This is pi calculated: {testpi}")
        print(abs(numpy.pi - testpi)/numpy.pi)
    negative +=1
```

```python
# As each term is added, the error starts to go down. 
```
