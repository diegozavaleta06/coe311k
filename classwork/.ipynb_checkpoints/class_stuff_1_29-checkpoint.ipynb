```python
import math
import random


def pi_counter (num_samples):
    counter = 0
    for s in range(0,num_samples):
        value = random.uniform(-1,1)
        value2 = random.uniform(-1,1)
        distance = value*value + value2*value2
        if distance <= 1:
            counter = counter + 1
    print(4 * (counter/num_samples))
    

        
        
        
```
